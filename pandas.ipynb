DATA_FRAME

```python
personas = { 
    "peso": pd.Series([20, 51, 74, 6], ["filipo", "la kate", "miguel et", "jose"]),
    "edad": pd.Series([15, 35, 23, 3], ["filipo", "la kate", "miguel et", "jose"]),
    "hijos": pd.Series([0, 5, 1, 0], ["filipo", "la kate", "miguel et", "jose"]),
}

table = pd.DataFrame(personas)
table
```

```python
listaList = np.empty((1, 3, 4))
table2 = pd.DataFrame(listaList[0], columns=["c1", "c2", "c3", "c4"], index=["i1", "i2", "i3"])
table2
```
