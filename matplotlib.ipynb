```python
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)

plt.plot(x, y)
plt.title("segundo grafico")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
y2 = x.copy()**2
y3 = x.copy()+1

plt.plot(x, y2, 'b--', x, y3, 'g')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(x, y2, 'b--')
plt.subplot(1, 2, 2)
plt.plot(x, y3, 'g')
plt.show()
```
